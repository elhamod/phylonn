```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def extract_csv_gen_plot(csv_path):

    data = pd.read_csv(csv_path, index_col=0)
#     print(data.columns)
#     data.index.names = ['Name']
    g = sns.heatmap(data, yticklabels=data.index, xticklabels=data.columns)
#     g.set_xticklabels(g.get_xticklabels(), rotation=90)
#     g.set_xticklabels(data.index)
    plt.tight_layout()
    plt.show()


extract_csv_gen_plot("PhyloNN-phylocodes.csv")
```
